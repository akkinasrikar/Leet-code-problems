```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1,l2):
        l1val=self.getvals(l1,[])
        l2val=self.getvals(l2,[])
        c1,c2=len(l1val),len(l2val)
        if c1==c2==1:
            t=l1val[0]+l2val[0]
            if t>=10:
                l1.val=1
                l1.next=ListNode(t%10)
                return l1
            else:
                l1.val=t
                return l1
            
        if c1>c2:
            l2val=l2val+[0]*(c1-c2)
            #print(l2val)
            temp=l1
        elif c2>c1:
            l1val=l1val+[0]*(c2-c1)
            #print(l1val)
            temp=l2
        else:
            temp=l2
        #print(l1val,l2val)
        nw=[]
        pre=0
        for i in range(len(l1val)):
            #print(i)
            v=l1val[i]+l2val[i]+pre
            if v<10:
                nw.insert(0,v)
                pre=0
            else:
                nw.insert(0,v%10)
                pre=1
        c=0
        while temp:
            temp.val=nw[c]
            temp=temp.next
            c+=1
 
        if c1>c2:
            if pre==1:
                q=ListNode(1)
                q.next=l1
                return q
            else:
                return l1
        else:
            if pre==1:
                q=ListNode(1)
                q.next=l2
                return q
            else:
                return l2
    def getvals(self,node,res):
        if node!=None:
            res.append(node.val)
            self.getvals(node.next,res)
        return res[::-1]
```

```python
class Solution:
    def addTwoNumbers(self, l1,l2):
        num=str(self.getvals(l1,[])+self.getvals(l2,[]))
        node=ListNode(num[0])
        temp=node
        for i in num[1:]:
            temp.next=ListNode(i)
            temp=temp.next
        return node
    def getvals(self,node,res):
        if node!=None:
            res.append(str(node.val))
            self.getvals(node.next,res)
        return int("".join(res))
```
